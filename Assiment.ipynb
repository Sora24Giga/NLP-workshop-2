```python
import pandas as pd
import string
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from ordered_set import OrderedSet
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

## Read

```python
data = pd.read_csv('spam.csv', encoding="iso-8859-1")

df = pd.DataFrame()
df['label'] = data['v1']
df['text'] = data['v2']
```

```python
df
```

## Clean data

```python
df['text'] = df['text'].str.strip()
```

```python
df['text']
```

```python
stop_dict = set(stopwords.words('English'))

def clean_data(text):
    d = text
    d = re.sub(r"[^a-zA-Z\s]","",d)
    d = d.lower()
    d = word_tokenize(d)
    d = list(OrderedSet(d) - stop_dict)
    d = ' '.join(d)
    return d

def tokenize(text):
    d = text
    d = text
    d = re.sub(r"[^a-zA-Z\s]","",d)
    d = d.lower()
    d = word_tokenize(d)
    d = list(OrderedSet(d) - stop_dict)
    return d
```

```python
df['text2'] = df['text'].apply(lambda x : clean_data(x))
```

```python
df['length'] = df['text2'].apply(lambda x : len(x))
```

## Label Encoder

```python
label_encoder = LabelEncoder()
unique = ''
```

```python
for i in df['text2']:
    unique = unique + ' ' + i
unique = word_tokenize(unique)
unique = np.unique(unique)
label_encoder.fit(unique)
def label(x):
    text = x
    text = word_tokenize(text)
    return label_encoder.transform(text)
```

```python
df['LabelEncoder'] = df['text2'].apply(lambda x: label(x))
```
